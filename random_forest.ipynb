```python
# 데이터 불러오기
import urllib.request

#데이터 로드 및 전처리
import re
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

# 데이터 시각화
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt", filename="ratings_train.txt")
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings_test.txt", filename="ratings_test.txt")
```

```python
train_data = pd.read_table('ratings_train.txt') # 훈련 데이터
test_data = pd.read_table('ratings_test.txt')   # 테스트 데이터
```

```python
train_data['document'].nunique(), train_data['label'].nunique()  # 중복이 아닌 샘플의 개수
```

```python
train_data.drop_duplicates(subset = ['document'], inplace=True) # document 열에서 중복인 내용이 있다면 제거
```

```python
print(train_data.isnull().values.any())  # null 값이 존재하는지 확인
```

```python
train_data = train_data.dropna(how = 'any')   # null 값이 존재하는 행 제거
print(train_data.isnull().values.any())       # null 값이 존재하는지 확인
```

```python
train_data['document'] = train_data['document'].str.replace("[^ㄱ-ㅎ ㅏ-ㅣ 가-힣 ]","")
# 한글과 공백을 제외하고 모두 제거
train_data[:5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer   # sklearn에서 제공하는 벡터화 시스템

vectorizer = CountVectorizer(analyzer = 'word', max_features=5000)
train_data_features = vectorizer.fit_transform(train_data['document'])
y = train_data['label']
```

```python
train_input, eval_input, train_label, eval_label = train_test_split(train_data_features, y, 
                                                                    test_size=0.2, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier

# 랜덤 포레스트 분류기에 100개의 의사결정 트리를 사용
forest = RandomForestClassifier(n_estimators=100)

# 단어 묶음을 벡터화한 데이터와 정답 데이터를 가지고 학습을 시작한다
forest.fit(train_input, train_label)
```

```python
print("Accuracy: %f" %forest.score(eval_input, eval_label))
```
